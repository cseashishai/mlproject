### 3.8 Adding columns for "Total Score" and "Average"

```python
df['total_score'] = df['math_score']+df['reading_score']+df['writing_score']
df['average'] = df['total_score']/3
```

```python
df.head()
```

## Now we chek the Which student what score performe

```python
## Full Marks how  much student got it 
read_full = df[df['reading_score']==100]['average'].count()
write_full = df[df['writing_score']==100]['average'].count()
math_full = df[df['math_score']==100]['average'].count()

print(f"Number of Student full marks obtain in Math :  {math_full}")
print(f"Number of Student full marks obtain in Writen :  {write_full}")
print(f"Number of Student full marks obtain in  Reading :  {read_full}")
```

```python
## Less then marks got student
read_less = (df['reading_score']<=20).sum()
write_less = (df['writing_score']<=20).sum()
math_less = (df['math_score']<=20).sum()

print(f"Number of Student less_20 marks obtain in Math :  {math_less}")
print(f"Number of Student less_20 obtain in Writen :  {write_less}")
print(f"Number of Student less_20 obtain in  Reading :  {read_less}")
```

#####  Insights
 - From above values we get students have performed the worst in Maths 
 - Best performance is in reading section

#### BIVARIATE ANALYSIS ( Is gender has any impact on student's performance ? ) 

```python
gender_group = df.groupby('gender', as_index=False).mean(numeric_only=True)
gender_group
```

```python
plt.figure(figsize=(10, 8))

X = ['Total Average','Math Average']

# Gender ko temporarily index bana do taaki .loc se direct value mil jaye
temp = gender_group.set_index('gender')

female_scores = [temp.loc['female', 'average'], temp.loc['female', 'math_score']]
male_scores   = [temp.loc['male', 'average'],   temp.loc['male', 'math_score']]

X_axis = np.arange(len(X))
  
plt.bar(X_axis - 0.2, male_scores, 0.4, label = 'Male') # left side space use -0.2
plt.bar(X_axis + 0.2, female_scores, 0.4, label = 'Female') # right side to use +0.2
  
plt.xticks(X_axis, X)
plt.ylabel("Marks")
plt.title("Total average v/s Math average marks of both the genders", fontweight='bold')
plt.legend()
plt.show()
```

#### Insights 
- On an average females have a better overall score than men.
- whereas males have scored higher in Maths.

#### BIVARIATE ANALYSIS ( Is Race/Ehnicity has any impact on student's performance ? )

```python
group_data2=df.groupby('race_ethnicity')

f,ax=plt.subplots(1,3,figsize=(20,8))

sns.barplot(x=group_data2['math_score'].mean().index,
            y=group_data2['math_score'].mean().values,
            palette='mako',
            ax=ax[0])
ax[0].set_title('Math Score',color='#005ce6',size=20)

for container in ax[0].containers:
    ax[0].bar_label(container,color='black',size=15)

sns.barplot(x=group_data2['reading_score'].mean().index,
            y=group_data2['reading_score'].mean().values,
            palette='flare',
            ax=ax[1])
ax[1].set_title("Reading Score",color='#005ce6',size=20)

for container in ax[1].containers:
    ax[1].bar_label(container,color='black',size=15)

sns.barplot(x=group_data2['writing_score'].mean().index,
            y=group_data2['writing_score'].mean().values,
            palette='coolwarm',
            ax =ax[2])

ax[2].set_title('Writing Score',color='#005ce6',size=20)

for container in ax[2].containers:
    ax[2].bar_label(container,color='black',size=15)

```

#### Insights 
- Group E students have scored the highest marks. 
- Group A students have scored the lowest marks. 
- Students from a lower Socioeconomic status have a lower avg in all course subjects
